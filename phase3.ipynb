# Advance operations with business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
print("=== zomato sales analysis ===")
print("\n Sales data shape: ", sales_data.shape)
print("\n Sample data for first 3 resturants: \n",sales_data[0:3])
print("\n Sample data for first 3 resturants: \n",sales_data[:3]) #both methods are correct
print("\n Sample data for first to end resturants: \n",sales_data[:, 1:]) # array(row, column) --> : --> row part(all rows),  :1 --> column partt(col from 1 to end(skips 0th column means ids))
```

# total sales per year


```python
print(np.sum(sales_data, axis=1)) #sum of rows(including ids)
print(np.sum(sales_data, axis=0)) #sum of cols(including ids)
yrs_total=(np.sum(sales_data[:, 1:], axis=0)) # here not include the 0th col(id col)
print(yrs_total)
```

# min sales per resturant

```python
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)
```

# max sales per yr

```python
max_sales = np.max(sales_data[:, 1:], axis=0)
#axis=0 for columns and axis=1 for rows
print(max_sales)
```

# avg sales per resturant

```python
avg_sales = np.average(sales_data[:, 1:], axis=1)
print(avg_sales)
```

# avg sales per yr

```python
avg_sales = np.average(sales_data[:, 1:], axis=0)
print(avg_sales)
```

# cumulative sales per yr

```python
cum_sales=np.cumsum(sales_data[:, 1:], axis=1)
print(cum_sales)
plt.figure(figsize=(8, 6))
plt.plot(np.mean(cum_sales, axis=0))
plt.title("Avg cumulative sales across all resturants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

# different vector operations

### (1) Vector Magnitude (Length) -> |v| = √(x² + y² + z²):
### np.linalg.norm(vector1)

### (2) Angle Between Two Vectors (CORE CONCEPT):
### angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))

### this is based on the formula, cos(θ)=v1​⋅v2​​/ |v1||v2|




```python
vector1=np.array([1,2,3])
vector2=np.array([4,5,6])

print("Vector addition: ", vector1 + vector2)
print("\nvector multiplication: ", vector1*vector2)
print("\n Dot product ", np.dot(vector1, vector2))


angle=np.arccos(np.dot(vector1, vector2)/(np.linalg.norm(vector1)* np.linalg.norm(vector2)))
print("Angle b/w vector 1 and vector 2: ",angle)
```
